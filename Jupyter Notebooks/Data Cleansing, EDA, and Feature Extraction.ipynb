```python
# Data Cleansing
```

```python
# Getting the Height (In) column

def getHeightInches(height):
    if isinstance(height, float):
        return None
    if len(height) == 4:
        return (12*int(height[0]) + int(height[2:]))
    else:
        return (12*int(height[0]) + int(height[2]))
    return None

dfAll['Height (In)'] = [getHeightInches(h) for h in dfAll['height']]
```

```python
print('Min Year:', dfAll['active_from'].min())
print('Mean Year:', dfAll['active_from'].mean())
print('Median Year:', dfAll['active_from'].median())
print('Max Year:', dfAll['active_from'].max())
```

```python
# Getting Actual Active_From column 
# active_from column in original dataset has values 1 year later than what the correct values should be

def getActualActiveFrom(year):
    return year-1

dfAll['Active_From'] = [getActualActiveFrom(y) for y in dfAll['active_from']]
```

```python
# Getting the Years_In_NBA column

def getYearsInNBA(year1, year2):
    return year2-year1

dfAll['Years_In_NBA'] = [getYearsInNBA(y1, y2) for (y1,y2) in zip(dfAll['Active_From'], dfAll['active_to'])]
```

```python
# Some players in their NCAA stats have high a 3-point percentage; however, they only shot a few 3-pointers (like David Robinson)
# David Robinson's NCAA 3-point percentage was 100%; however, he only made and attempted 1 3-pointer
# Thus, the NCAA__3ptpctpg column will multiply a player's NCAA 3ptpct by their NCAA 3ptpg to determine if they were accurate 3-point shooters who shot a good amount of 3-pointers in their NCAA careers

def getNCAAThreePtPctPg(pct,pg):
    return pct*pg
dfNCAA_1980['NCAA__3ptpctpg'] = [getNCAAThreePtPctPg(percentage, pergame) for (percentage, pergame) in zip(dfNCAA_1980['NCAA__3ptpct'], dfNCAA_1980['NCAA__3ptpg'])]
```
